```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import re
import csv
import pandas as pd
import numpy as np

driver = webdriver.Chrome()

# Scrape each category at a time
# Go to the category page that we want to scrape                
category = 'moisturizing-cream-oils-mists'
# category = 'facial-treatments'
# category = 'eye-treatment-dark-circle-treatment'
# category = 'face-mask'
# category = 'sunscreen-sun-protection'
# category = 'cleanser'
driver.get(f"https://www.sephora.com/shop/{category}") 

mosturizers = []
# faceTreatments = []
# eyeTreatments = []
# faceMasks = []
# sunCare = []
# cleansers = []

# Page index used to keep track of where we are.
page = 1

# 14 pages of moisturizers, which include creams, oils, and mists
# 10 pages of facial treatments, 4 pages of eye treatments, 
# 5 pages of face masks, 3 pages of sunscreen, 9 pages of cleansers
while page <= 14:
    try:
        print("Scraping Page number " + str(page))
        page = page + 1
        
        # Scroll down in segments to allow all product links to show
        body_height = 7000
        start = 0
        while start < body_height:
            driver.execute_script("window.scrollTo(%i, %i);" % (start, start+1000))
            start += 1000
            time.sleep(5)
            
        # Get link to each product on page
        links = [link.get_attribute('href') for link in driver.find_elements_by_xpath('//div[@class="css-12egk0t"]/a')]
        # ===print(len(links))
          
        # Go into each product page and scrape needed data
        i = 0
        for link in links:
            driver.get(link)
            time.sleep(3)
            product_dict = {}
            i += 1
            #print(i)
            # Close popup window
            try:
#                 wait = WebDriverWait(driver, 5)
#                 closebtn = wait.until(EC.visibility_of_element_located((By.XPATH, '//button[@class="css-1kna575"]')))[0]
                closebtn = driver.find_elements_by_xpath('//button[@class="css-1kna575"]')[0]
                closebtn.click()
            except:
                pass
            
            brand = driver.find_elements_by_class_name("css-57kn72")[0].text
#             print(brand)
            product_name = driver.find_elements_by_xpath('//h1/span')[0].text
#             print(product_name)
            price = driver.find_elements_by_xpath('//div[@class="css-1865ad6 e65zztl0"]/span')[0].text
#             print(price)
            
            # Try to scrape size of product from two different locations on page
            try:
                size = driver.find_element_by_xpath('//div/span[@class="css-ta42ek e65zztl0"]').text
            except:
                size = driver.find_element_by_xpath('//div[@class="css-128n72s e65zztl0"]').text
                pass
            
            ingredients_callouts = driver.find_element_by_xpath('//div[@id="tabpanel0"]/div').text
#             print(ingredients_callouts)

            # Wait for ingredients tab to be clickable
            wait = WebDriverWait(driver, 12)
            ingredients_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//button[@id="tab2"]')))
            ingredients_button.click()
            # Wait for ingredients text to all be visible 
            wait = WebDriverWait(driver, 10)
            product_ingredients = wait.until(EC.visibility_of_element_located((By.ID, 'tabpanel2'))).text
#             print(product_ingredients)
            
            star_review = driver.find_element_by_xpath('//div[@class="css-jp4jy6"]').get_attribute('aria-label')
#             print(star_review)
        
            num_reviews = driver.find_element_by_xpath('//div/a/span[@class="css-2rg6q7"]').text
#             print(num_reviews)
            
            # Write values to product_dict
            product_dict['category'] = category
            product_dict['brand'] = brand
            product_dict['product_name'] = product_name
            product_dict['price'] = price
            product_dict['size'] = size
            product_dict['star_review'] = star_review
            product_dict['num_reviews'] = num_reviews
            product_dict['ingredients_callouts'] = ingredients_callouts
            product_dict['ingredients'] = product_ingredients
            
            mosturizers.append(product_dict)
#             faceTreatments.append(product_dict)
#             eyeTreatments.append(product_dict)
#             faceMasks.append(product_dict)
#             sunCare.append(product_dict)
#             cleansers.append(product_dict)
            

        # Go to next page
        driver.get(f"https://www.sephora.com/shop/moisturizing-cream-oils-mists?currentPage={page}")
        #driver.get(f"https://www.sephora.com/shop/facial-treatments?currentPage={page}")
        #driver.get(f"https://www.sephora.com/shop/eye-treatment-dark-circle-treatment?currentPage={page}")
        #driver.get(f"https://www.sephora.com/shop/face-mask?currentPage={page}")
        #driver.get(f"https://www.sephora.com/shop/sunscreen-sun-protection?currentPage={page}")
        #driver.get(f"https://www.sephora.com/shop/face-wash-facial-cleanser?currentPage={page}")
        #driver.get(f"https://www.sephora.com/shop/exfoliating-scrub-exfoliator?currentPage={page}")
        time.sleep(10)

    except Exception as e:
        print(e)
        driver.close()
        break
```
